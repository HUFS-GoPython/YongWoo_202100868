 ### 반복문
 - while, for loop
 
 #### while loop
 - 어떤 조건이 만족하는 동안 명령 블록을 수행한다.
 - if + for loop (조건문 + 반복문)
 - 해당 조건이 거짓이면 멈춘다. ==> 언제 멈추는지 그 시점을 정확히 파악해야함
 
 #### for vs. while
 - for: 반복 횟수가 정해짐 -> 정확함, 변하지 않음
     - 예: 성적 산출(모든 학생의 수 만큼 성적을 산출할 때 까지 반복)
 - while: 반복횟수 정해지지 않음, 반복횟수가 중요하지 않음 ->어떤 조건을 만족하는 경우 중단할 때 사용 
     - 예: 가위바위보(결과가 나올 때 까지 반복)

```python
#for loop

for i in range(10):
    print(i)
```

```python
#위의 for loop을 while loop로 변환!!
# 초기화와 변화가 가장 큰 차이점

j = 0 # 초기화
while j < 10:
    print(j)
    j += 1 # 변화
```

```python
#for loop

drinks =['soju','beer','rum','wine']
for drink in drinks:
    print(drink)
```

```python
#while loop !초기화 및 변화!

i = 0
while i < len(drinks):
    print(drinks[i])
    i += 1
```

```python
# 변화하는 위치에 따라서 출력값이 아예 달라지므로 주의!

i = 0
while i < len(drinks):
    i += 1
    print(drinks[i])
# 이 경우에는 인덱스 값이 drinks의 최대 인덱스보다 커서 오류가 발생
```

### 반복문 제어
- 되도록 사용하지 않는다. 언제 어디서 종료되고 오류가 발생할지 모름
- break: 논리적으로 반복을 종료시킬때
- else : 어떤 조건이 완전히 끝났을 때 한번 더 실행함. break 시에 실행 안됨 ==>언제 종료되었는지 확인할 때, for, while과 같은 위치에 들여씀
- continue: 특정조건을 건너뛰고 다음 조건 실행

```python
j = 0 # 초기화
while j < 10:
    print(j)
    if j == 6: break #반복을 종료하고 while loop을 빠져나옴
    j += 1 # 변화
```

```python
# break 위치에 따라서 다른 결과가 나옴
j = 0
while j < 10:
    if j == 6: break
    print(j)
    j += 1
```

```python
for i in range(10):
    if i == 6: break
    print(i)
```

```python
# break 되었으므로 else는 실행 X
k = 0
while k < 10:
    print(k)
    if k == 3: break
    k += 1
else: 
    print("ended")
```

```python
#break 되지 않았으므로 else 실행
k = 0
while k < 10:
    print(k)
    if k == 15: break
    k += 1
else: 
    print("ended")
```

```python
#for loop else

for i in range(5):
    if i == 6: break
    print(i)
else: print("ended")
```

```python
#continue 조건이 만족되면 넘어간다

for i in range(5):
    if i == 3: continue
    print(i)
# 3일 때는 출력 X
```

```python
#각 파일마다 word, character, sents 개수 출력하기

for i in gutenberg.fileids():
    words = len(gutenberg.words(fileids = i)) #words 는 텍스트의 단어들 출력
    sents = len(gutenberg.sents(fileids = i)) #sents 는 텍스트의 문장 출력
    chars = len(gutenberg.raw(fileids = i)) #raw 는 텍스트의 글자수 출력
    print(f'{i}: 글자수 {chars}개, 단어수 {words}개, 문장수 {sents}개')
```

### Gutenberg quiz
- 1. shakespeare-caesar.txt의 중복된 단어를 거르고 각단어의 길이를 출력
- 2. chesterton-thursday.txt에서 길이가 4자 이하인 알파벳 중 ie를 포함하는 단어를 소문자로 출력하고 개수를 세어라
- 3. bryant-stories.txt 에서 ch로 끝나는 단어를 출력하는데, 중복없이, 첫글자(모두 소문자)의 끝 ch만 남기고 가운데 글자들은 모두 - 로 가려라

```python
word_len = [len(word) for word in set(gutenberg.words(fileids = 'shakespeare-caesar.txt'))]
word_len
```

```python
#2. chesterton-thursday.txt에서 길이가 4자 이하인 알파벳 중 ie를 포함하는 단어를 소문자로 출력하고 개수를 세어라
# 난 중복된걸 다 삭제하고 개수를 셈
word_list = set([word.lower() for word in gutenberg.words(fileids='chesterton-thursday.txt')
            if len(word) < 4 and 'ie' in word])
print(word_list, len(word_list))
```

```python
#3.bryant-stories.txt 에서 ch로 끝나는 단어를 출력하는데, 중복없이, 첫글자(모두 소문자)의 끝 ch만 남기고 가운데 글자들은 모두 - 로 가려라
word_list = [(word[0] + '-' * (len(word) - 3) + 'ch') for word in set(map(lambda x: x.lower(),gutenberg.words(fileids = 
                'bryant-stories.txt'))) if word.endswith('ch')]
# list 속 단어들을 소문자로 모두 변환후 중복된 것을 제거함 그렇지 않고 set()하면 Each 와 each 같이 대소만 다른 문자를 구별 못함
len(word_list)
```

### Brown
- categories()

```python
brown = nltk.corpus.brown
```

```python
brown.categories() #카테고리만 출력가능
```

```python
fdist = nltk.FreqDist(w.lower() for w in brown.words(categories='adventure'))
#adventure 카테고리의 fdist
```

```python
fdist['men'] 
#adventure 카테고리에서의 men의 출력수
```

```python
#modal verbs
modals = ['can','could','will','would','may','might']
for verb in modals:
    print(verb, fdist[verb])
# 각 조동사들의 출력 빈도 수 
```

### conditional frequency distribution
- 조건에 따른 fdist

```python
cfd = nltk.ConditionalFreqDist(
    #각 장르에 따른 frequency distribution을 만든다 () 형태로
    (genre,word.lower())
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)
```

```python
cfd['romance'].freq('like')
#로맨스 카테고리에서의 'like'
```

```python
cfd['romance']['love'] 
#로맨스 카테고리에서의 love 출현빈도수
```

```python
cfd['romance'].plot(50,cumulative = True)
#로맨스 카테고리에서의 빈번한 50개의 단어를 이미지로 출력
```
