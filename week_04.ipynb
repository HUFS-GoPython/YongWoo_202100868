### Frequency Distribution
- 단어의 빈도수를 활용할 수 있는 많은 메소드 제공
- 인간 언어 -> 컴퓨터가 인식
    - 전체단어수
    - 특정 빈도수
- FreqDist() 는 단어들과 출력된 횟수를 딕셔너리 형태로 반환
- .most_common(수): 가장 빈번하게 나타난 단어들을 횟수와 함께 출력
- FreqDist()['str']: 딕셔너리의 'str'의 value값을 출력
- .freq('str') : 전체 텍스트에서 스트링이 얼마나 자주 나타나는지 비율을 출력
- .plot(수,cumulative = True or False): 자주 나타나는 단어들의 수를 이미지로 출력, cumulative를 통해 누적으로 표현 가능
- .hapaxes(): 한번만 나타나는 단어들을 리스트로 출력(중복되지 않지만 대소문자를 구분함)

```python
text1 = book.text1
fdist = nltk.FreqDist(text1)
fdist # 딕셔너리형태로 반환
```

```python
fdist.most_common(10) #가장 빈번한 10개의 단어를 횟수와 함께 출력
```

```python
fdist['whale'], text1.count('whale') #'whale'의 횟수를 출력
```

```python
fdist.freq('whale') #'whale'이 나타난 비율을 출력
```

```python
fdist.plot(10,cumulative=True) #빈번하게 나타난 10개의 단어들을 이미지로 출력 누적해서
```

```python
#1번 나타나는 단어들
len(fdist.hapaxes()) # 중복 안됨, 대소문자 구분됨
```

#### list comprehension
- 지능형 리스트
- 주어진 리스트를 포함하는 새로운 리스트를 형성
- list + for_loop 한줄로 해결하는 방식
- [A for A' in B]
    - B에 있는 A'에서 A를 가진 리스트를 반환
    - B: 기존의 리스트
    - A': B의 자료 형태
    - A: 새로운 리스트의 요소
- 조건+: [A for A' in B if <조건식>]
    - [A for A' in B if <조건식> and <조건식>,...]

```python
[d for d in drinks]
```

```python
[d[0] for d in drinks] #첫번째 단어들만 골라서 리스트 만들수 있음
```

```python
#text1 알파벳 수가 10 이상인 단어
[word for word in text1 if len(word) >= 10]
```

```python
# text4에서 'gnt'를 포함하는 단어 반환
[word for word in set(book.text4) if 'gnt' in word]
#set([word for word in book.text4 if 'gnt' in word]) 중복을 피하기 위해서
```
